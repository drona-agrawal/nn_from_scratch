```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
```

```python
X = np.linspace(-5, 5, 100)
X.reshape(-1,1)
```

```python
y = (X > 0).astype(float)
y
```

```python
w1 = np.random.randn()
b1 = 0.0

w2 = np.random.randn()
b2 = 0.0
```

```python
lr = 0.1
epochs = 2000
```

```python
def sigmoid(z):
    return 1/(1 + np.exp(-z))
```

```python
losses = []

for epoch in range(epochs):

    total_loss = 0

    for i in range(len(X)):
        x = X[i]
        target = y[i]

        # Forward Pass
        z1 = w1 * x + b1
        a1 = sigmoid(z1)

        z2 = w2 * a1 + b2
        a2 = sigmoid(z2)

        # Loss
        loss = -(target*np.log(a2 + 1e-8) + (1-target)*np.log(1-a2 + 1e-8))
        total_loss = total_loss + loss

        # Output layer
        dz2 = a2 - target
        dw2 = dz2 * a1
        db2 = dz2

        # Hidden layer
        dz1 = dz2 * w2 * a1 * (1 - a1)
        dw1 = dz1 * x
        db1 = dz1

        # Update
        w2 -= lr * dw2
        b2 -= lr * db2

        w1 -= lr * dw1
        b1 -= lr * db1

    print(total_loss)
    losses.append(total_loss)
```

```python
preds = []

for i in range(len(X)):
    x = X[i]
    a1 = sigmoid(w1 * x + b1)
    a2 = sigmoid(w2 * a1 + b2)
    preds.append(a2)

plt.scatter(X, y, label="True")
plt.plot(X, preds, color='red', label="Prediction")
plt.legend()
plt.title("1-1-1 Neural Network")
plt.show()

plt.plot(losses)
plt.title("Loss")
plt.show()
```
